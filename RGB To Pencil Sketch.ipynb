```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def convert_to_pencil_sketch(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Invert the grayscale image
    inverted_image = cv2.bitwise_not(gray_image)

    # Create a blurred version of the inverted image
    blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)

    # Divide the grayscale image by the inverted blurry image
    pencil_sketch = cv2.divide(gray_image, blurred_image, scale=256.0)

    return pencil_sketch

# Read the image in RGB format
image = cv2.imread("C:/Users/HP/Desktop/IRMAN.jpg")

# Convert the image from BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the original image
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image)
plt.axis('off')

# Convert the image to pencil sketch
pencil_sketch = convert_to_pencil_sketch(image)

# Display the pencil sketch image
plt.subplot(1, 2, 2)
plt.title('Pencil Sketch')
plt.imshow(pencil_sketch, cmap='gray')
plt.axis('off')

# Adjust the subplot spacing
plt.tight_layout()

# Show the figure with both images
plt.show()
```
